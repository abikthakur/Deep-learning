```python
import tensorflow as tf
```

```python
print(tf.__version__)
```

```python
print(tf.config.list_physical_devices('GPU'))
```

```python
a = tf.constant(5)
print(a)
v = tf.constant([1,2,3,4])
print(v)
```

```python
m = tf.constant([[1,2],[3,4]])
print(m)
```

```python
t3 = tf.constant([[[1,2],[3,4]],[[5,6],[7,8]]])
print(t3 ,end =" ")
```

```python
x = tf.constant([1, 2, 3], dtype=tf.float32)
print(x)
```

```python
a = tf.constant([10,20,30])
b = tf.constant([50,60,70])
print(a+b)
print(a*b)
print(a -b)
print(b-a)
print(a /b)
print(a**b)
```

```python
v1 = tf.constant([1, 2, 3], dtype=tf.float32)
v2 = tf.constant([4, 5, 6], dtype=tf.float32)

print("\nVector Addition:", v1 + v2)
print("Vector Subtraction:", v1 - v2)
print("Vector Element-wise Multiplication:", v1 * v2)
print("Vector Element-wise Division:", v1 / v2)
print("Vector Power:", tf.pow(v1, v2))
```

```python
m1 = tf.constant([[1, 2],
                  [3, 4]], dtype=tf.float32)

m2 = tf.constant([[5, 6],
                  [7, 8]], dtype=tf.float32)

print("\nMatrix Addition:\n", m1 + m2)
print("Matrix Subtraction:\n", m1 - m2)
print("Matrix Element-wise Multiplication:\n", m1 * m2)
print("Matrix Element-wise Division:\n", m1 / m2)
print("Matrix Power:\n", tf.pow(m1, m2))
```

```python
z = tf.zeros((2, 3))
print(z)
```

```python
o = tf.ones((3, 2))
print(o)
```

```python
f = tf.fill((2, 2), 9)
print(f)
```

```python
r = tf.range(1, 10, 2)
print(r)
```
